# Preprocessing

```python
drop_cols = ['IDX', '카테고리', 'QA번호', '가격', '상담번호', '상담내순번', '날짜', '시간', '사람']
processed_data = restaurant.drop(drop_cols, axis=1)
processed_data.head()
```

```python
processed_data['상품명'] = processed_data['상품명'].str.replace(" ", "")
processed_data['상품명1'] = processed_data['상품명'].str.split('|').str[0]
processed_data['상품명2'] = processed_data['상품명'].str.split('|').str[1]
processed_data['상품명3'] = processed_data['상품명'].str.split('|').str[2]
processed_data['상품명4'] = processed_data['상품명'].str.split('|').str[3]
processed_data['상품명5'] = processed_data['상품명'].str.split('|').str[4]
processed_data['상품명6'] = processed_data['상품명'].str.split('|').str[5]
processed_data['상품명7'] = processed_data['상품명'].str.split('|').str[6]
processed_data['상품명8'] = processed_data['상품명'].str.split('|').str[7]
processed_data['상품명9'] = processed_data['상품명'].str.split('|').str[8]
processed_data['상품명10'] = processed_data['상품명'].str.split('|').str[9]
processed_data['상품명11'] = processed_data['상품명'].str.split('|').str[10]
processed_data['상품명12'] = processed_data['상품명'].str.split('|').str[11]
processed_data['상품명13'] = processed_data['상품명'].str.split('|').str[12]
processed_data['상품명14'] = processed_data['상품명'].str.split('|').str[13]
processed_data['상품명15'] = processed_data['상품명'].str.split('|').str[14]
processed_data['상품명16'] = processed_data['상품명'].str.split('|').str[15]
processed_data['상품명17'] = processed_data['상품명'].str.split('|').str[16]
processed_data['상품명18'] = processed_data['상품명'].str.split('|').str[17]
processed_data['상품명19'] = processed_data['상품명'].str.split('|').str[18]
processed_data['상품명20'] = processed_data['상품명'].str.split('|').str[19]
processed_data['상품명21'] = processed_data['상품명'].str.split('|').str[20]
processed_data['상품명22'] = processed_data['상품명'].str.split('|').str[21]

print(processed_data['상품명22'].value_counts())
processed_data
```

```python
hambuger = processed_data.drop('상품명', axis=1)

ham1 = hambuger[hambuger.상품명1 == '햄버거']
ham2 = hambuger[hambuger.상품명1 == '버거']
ham3 = hambuger[hambuger.상품명1 == '햄버거세트']
ham4 = hambuger[hambuger.상품명1 == '버거세트']
ham5 = hambuger[hambuger.상품명1 == '치킨버거']
ham6 = hambuger[hambuger.상품명1 == '불고기버거']
ham7 = hambuger[hambuger.상품명1 == '불고기버거세트']
ham8 = hambuger[hambuger.상품명1 == '치즈버거']
ham9 = hambuger[hambuger.상품명1 == '더블치즈버거']
ham10 = hambuger[hambuger.상품명1 == '상하이스파이시버거세트']
ham11 = hambuger[hambuger.상품명1 == '패티']
ham12 = hambuger[hambuger.상품명1 == '프렌치프라이']
ham13 = hambuger[hambuger.상품명1 == '빅맥']

hambuger = pd.concat([ham1, ham2, ham3, ham4, ham5, ham6, ham7, ham8, ham9, ham10, ham11, ham12, ham13])
hambuger
```
